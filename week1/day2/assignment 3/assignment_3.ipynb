#### group By with joins

```sql
--16
select e.emp_name, sum(s.amount) as total_sales from employees e
join Sales s
on e.emp_id=s.emp_id
group By e.emp_name;
```

```sql
--17
select e.emp_name, count(s.sales_id) as number_of_sales from employees e
join Sales s
on e.emp_id=s.emp_id
group By e.emp_name;
```

```sql
--18
select product, sum(amount) as total_sales_amount from Sales
group By product;
```

```sql
--19
select product, avg(amount) as avg_sales_amount from Sales
group By product;
```

```sql
--20
select e.emp_name, count(s.sales_id) as no_of_sales from employees e
join Sales s
on e.emp_id=s.emp_id
group By e.emp_name 
having no_of_sales>2;
```

#### Advanced group By

```sql
--21
select e.emp_name, sum(e.salary) as total_salary, sum(s.amount) as total_sales_amount from employees e
join Sales s
on e.emp_id=s.emp_id
group By e.emp_name;
```

```sql
--22
select e.emp_name, count(distinct s.product) as unique_product from employees e
join Sales s
on e.emp_id=s.emp_id
group By emp_name;
```

```sql
--23
select e.emp_name, max(s.amount) as highest_sales_amount from employees e
join Sales s
on e.emp_id=s.emp_id
group By emp_name;
```

```sql
--24
select product, sum(amount) as total_sales_amount from Sales 
group By product
having total_sales_amount>50000;
```

```sql
--25
select e.department, avg(s.amount) as highest_average from employees e
join Sales s 
on e.emp_id=s.emp_id
group By e.department
order By highest_average Desc
limit 1;
```
