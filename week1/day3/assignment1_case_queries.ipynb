#### Problem 1: Salary Hike Based on Experience and Performance

```sql
select emp_name,emp_id,department,salary,experience,performance_rating,
case when performance_rating = 'A' and experience>=8 then salary*1.2
when performance_rating = 'B' and experience>=5 then salary*1.15
when performance_rating = 'C' then salary*1.0
else salary*1.1 end as new_salary
from Employeee
```

#### Problem 2: Bonus Calculation Based on Department and Performance

```sql
select emp_name,emp_id,department,salary,
case 
  when department='finance' then 
    case when performance_rating = 'A' then salary*1.2 
    when performance_rating = 'B' then salary*1.15
    when performance_rating = 'C' then salary*1.05 end
  when department='engineering' then 
    case when performance_rating = 'A' then salary*1.18 
    when performance_rating = 'B' then salary*1.12
    when performance_rating = 'C' then salary*1.03 end
else salary*1.1 end as bonus_salary
from Employeee
```

#### Problem 3: Categorizing Employees by Salary Range and Performance

```sql
select salary,performance_rating,
case 
when performance_rating='A' and salary>=80000 then 'High performer'
when performance_rating='B' and salary>=50000 and salary<80000 then 'Mid performer'
when performance_rating='C' and salary<50000 then 'Low performer'
end as performance
from employeee;
```

#### Problem 4: Risk Assessment Based on Experience and Department

```sql
select experience,department,
case 
  when department='HR'then 
  case 
    when experience<=5 then 'High Risk'
    when experience >=5 then 'Low risk'
  end
 
  when department in('Engineering','Finance')then 
  case
    when experience>=5 then 'Low Risk'
    when experience <=5 then 'Medium Risk'
  end
end as Risk_Assessment
from employeee
```

#### nested case and when

#### Problem 1: Nested CASE for Performance and Salary Hike Based on Multiple Criteria

```sql
select emp_name,emp_id,department,salary,
case 
  when performance_rating='A' then 
    case when salary>=80000 or experience>=5 then salary*1.25 
    when salary between 50000 and 80000 or experience>=5 then salary*1.15 
    else salary*1.2 end
  when performance_rating='B' then 
    case when experience>=5 then salary*1.12
    else salary*1.1 end
else salary*1.0 end as salary_hike
from Employeee
```

#### 2.Department and Performance

```sql
select department,performance_rating,experience,salary,
case 
  when department='Finance' then 
    case 
      when  performance_rating='A' and experience>10 then salary*1.25
      else salary*1.20
    end 
  when department='HR' then 
    case 
      when performance_rating='B' and experience>5 then salary*1.15
      else salary*1.10
    end
  else salary*1.10
end as bonus
from employeee;
```

#### Problem 3: Nested CASE for Employee Categorization Based on Salary, Performance, and ExperienceScenario

```sql
select salary,performance_rating,experience,
case 
  when salary>70000 then 
    case 
      when performance_rating='A' and experience>8 then 'Top Performer'
      when experience<8 then 'Mid Performer'
    end
  when salary between 50000 and 70000 then 
    case 
      when performance_rating='A'then 'Rising Star'
      else 'Average Performer'
    end 
  when salary<50000 then 'Low Performer'
end as Label
from employeee;
```

#### 4.Nested CASE for Tax Bracket Based on Salary and Experience

```sql
select *,
case 
when salary>=90000 then 
  case 
    when experience>=10 then '35%'
    else '30%'
  end
when salary between 60000 and 90000 then 
  case 
    when experience>=5 then '25%'
    else '5%'
  end
when salary<=60000 then '15%'
end as Tax_Rate
from employeee;
```

#### Problem 5: Nested CASE for Promotion Eligibility Based on Performance, Salary, and Experience

```sql
select salary,performance_rating,experience,
case 
  when performance_rating='A' then 
    case 
      when salary>75000 and experience>7 then 'Eligible for Senior Role'
      else 'Eligible for junior Role'
    end 
  when performance_rating='B' then 
    case 
      when experience>5 then 'Eligible for Consideration'
      else 'Not Eligible'
    end
  else 'Not Eligible'
  end as promotion_eligibility
  from employeee;
```
