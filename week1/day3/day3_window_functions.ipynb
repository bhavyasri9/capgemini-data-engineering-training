#### queries on row_number()

```sql
--1
select emp_id, emp_name,salary, row_number() over(order by salary desc) as row_number from emp;
```

```sql
--2
select emp_id, emp_name,salary, row_number() over(partition by department order by salary desc) as row_number from emp;
```

```sql
--3
select emp_id, emp_name, join_date, row_number() over(order by join_date desc) as row_number from emp;
```

```sql
--4
select emp_id, emp_name, join_date, row_number() over(partition by department order by join_date) as row_number from emp;
```
