### loading csv file

```python
df=spark.read.csv("/Volumes/workspace/default/day1/C1.csv",header=True,inferSchema=True)
df.show()
```

### handling 'blank'/'Blank' values


```python
from pyspark.sql.functions import *
df=df.replace(['blank','Blank'],None)
df.display()
```

### fixing country column

```python
df=df.withColumn("Country", upper(col("Country")))
df.display()
```
