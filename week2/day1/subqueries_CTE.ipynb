```sql
CREATE TABLE Employeess (
    emp_id INT PRIMARY KEY,
    name VARCHAR(50),
    department_id INT,
    salary INT,
    manager_id INT
);
```

```sql
drop table if exists Departments;

create table Departments (
    dept_id int primary key,
    dept_name varchar(50)
);
```

```sql
drop table if exists Orders;
CREATE TABLE Orders (
    order_id INT PRIMARY KEY,
    emp_id INT,
    order_amount INT,
    order_date DATE
);
```

### sub queries

```sql
--1
select name, salary from Employeess where salary > (select avg(salary) from Employeess);
```

```sql
--2
select e.name, d.dept_name from employeess e 
join departments d on e.department_id = d.dept_id
where d.dept_id=(select department_id from employeess where name='Alice');
```

```sql
--3
select name, salary from employeess where salary=(select min(salary) from employeess);
```

```sql
--4
select e.name, count(o.order_id) as order_count from employeess e join orders o on e.emp_id=o.emp_id group by e.name having order_count>=1;
```

```sql
--4.1
select e.* from employeess e where exists (select 1 from orders o where o.emp_id = e.emp_id);
```

```sql
--5
select * from employeess where salary > (select max(salary) from employeess where department_id = 102
);
```

### CTE (Common Table Expression)

```sql
--1
with highest_salary as(select * from employeess where salary>50000)
select * from highest_salary;
```

```sql
--2
with dept_avg as(select avg(salary)as avg_salary from Employeess group by department_id)
select * from dept_avg;
```

```sql
--3
with emp_with_dept as
(select d.dept_id,e.name from Departments d join Employeess e on d.dept_id=e.department_id)
select * from emp_with_dept;
```

```sql
--4
with emp_orders as
(select e.emp_id,sum(order_amount) as total_amount from Employeess e join Orders o on e.emp_id=o.emp_id group by e.emp_id)
select * from emp_orders;
```

```sql
--5
with dept_avg as (
    select department_id, avg(salary) as avg_salary
    from Employeess
    group by department_id
)
select e.emp_id, e.name, e.salary, d.dept_name
from Employeess e
join dept_avg d
on e.department_id = d.department_id
join Departments d
on e.department_id = d.dept_id
where e.salary > d.avg_salary;
```
